# Project 2: Ames House Price Prediction

## Introduction

This project aims to create and refine a regression model that predicts the price of a house at sale at the Ames housing area. Ames is a city in Iowa, United States and is best known as the home of Iowa State University.

To create this regression model, we are given the Ames Housing Dataset, which contains several columns of different features relating to houses. Examples of such features are Garage Area, Kitchen Quality etc.

## Problem Statement

Our goal is to determine what are the most important features in determining the price of the house. In addition, by using our regression model, home owners who are looking to sell can determine if the offered price is fair or not. If the price do not meet expectations, the home owners can then focus on upgrading these features to drive up the price of the house. We will be using root mean squared error to evaluate our models.

## <a name="sec_3"></a> Section 3: Dropping Columns/Features

In the current dataset, there are simply too many columns/features, some of which may not be useful.
Thus, we need to do a EDA (Exploratory Data Analysis) to determine the variables that should be considered dropping.

On initial inspection after reviewing the data description, there are several columns that stood out that we would look to drop first. For example, ID and PID are simply transaction numbers and should not be a factor in determing house prices. Sale Type should be dropped too as how a person chooses to pay for the house should not be a factor in determining house prices too, 

Also, though 'Mo Sold' and 'Yr Sold' do have an impact on the prices of houses sold, for example, during a financial crise or a housing bubble, we feel that they should be dropped as the prices are impacted for that period of time and should not be used as a basis for predicting future housing prices. 

We would also look to drop 'Year Built', 'Garage Yr Blt' and Year Remod/Add. This is because, though it might seem that a newer house and newer garage would fetch a higher price, it is most likely because this caused other factors to be better, like for example, kitchen quality or overall quality, after the remodeling. Thus, how old the house or garage is should be a minor factor compared to such qualities. 

Moreover, from the preivous section, we had already concluded to drop 'Pool QC', 'Misc Feature', 'Alley', 'Fence' and 'Lot Frontage'.


```python
train_df = train_df.drop(columns = ['Id','PID','Sale Type','Mo Sold',
                                    'Year Built','Yr Sold', 'Garage Yr Blt',
                                    'Year Remod/Add', 'Pool QC', 'Misc Feature',
                                   'Alley', 'Fence', 'Lot Frontage'])

# we will drop these 13 columns first
```

### 3.1 Dropping Numeric Columns

We will first focus on the Numeric Columns

We plot a correlation heatmap of the remaining numeric columns to dive deeper and see the correlation coefficents for them. At the same time, we can look out for variables with a high correlation to each other and consider removing some to reduce multicollinearity.

Listed below are our observations and action taken.

* Drop Garage Cars
    * Garage Cars and Garage Area have a high correlation (0.89). This makes sense as having a larger garage area likley means you have space to park more cars. Both Garage Cars and Garage Area have the same correlation in regards to SalePrice (0.65). We will drop Garage Cars as we feel Garage Area is a more descriptive and informative.


* Drop TotTms AbvGrd 
    * TotRms AbvGrd and Gr Living Area have a high correlation to each other (0.9). This makes sense as having a larger living area likley means that there are more rooms in the house. In regards to SalePrice, Gr Living Area (0.72) has a higher correlation than TotTms AbvGrd (0.51). Thus, we will be dropping TotRms AbvGrd
    
    
* Drop Numeric features with correlations < 0.3 or > -0.3
    * except for Bsmt Full Bath. between Bsmt Full Bath and Bsmt Half Bath, Bsmt Full Bath has a higher correlation of 0.28. though it is less than 0.3, we will keep it as we feel it is at least better to have a parameter related to basement bathrooms in our regression model.

```python
def value_counts_bool(df):
    features_to_drop = []
    for i in df.columns:
        a = df[i].value_counts(normalize=True)
        b = 0
        for name, perc in a.iteritems():
            if perc >= 0.8:
                b += 1
        if b ==1:
            features_to_drop.append(i)
    return features_to_drop
        
        
        
```
